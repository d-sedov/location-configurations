```python
############################### Libraries ######################################

import requests
from bs4 import BeautifulSoup
import pandas as pd
import os
import us
import json

################################################################################
```

```python
######################### Constants and settings ###############################

url_search = 'https://api.commercialexchange.com/api/cfra/ee/search/properties/_list?pageSize=100&sort=QUALITY%3Adesc&byProperty=true'
url_property = 'https://api.commercialexchange.com/api/cfra/ee/property/{}'

user_agent = 'Moilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.132 Safari/537.36'

output_folder_path = '/home/user/projects/urban/data/input/Rent/commercialexchange/retail'

form_data = {'pageNumber':1,
             'locations':[{'postalCode':'54016'}],
             'listedStatus':['AVAILABLE_LEASE']
            }

headers_search = {'x-api-key': '***REMOVED***',
                  'user-agent': user_agent,
                  'content-type': 'application/json',
                  'accept': 'application/json'
                 }

headers_property = {'x-api-key': '***REMOVED***',
                    'user-agent': user_agent,
                    'accept': 'application/json'
                   }

################################################################################
```

```python
session = requests.Session()
```

```python
result = session.post(url_search, data = json.dumps(form_data), headers = headers_search)
```

```python
result.json()
```

```python
properties = [x['data']['id'] for x in result.json()['results']]
```

```python
result_search = session.get(url_property.format(properties[3]), headers = headers_property)
```

```python
result_search.json()
```
